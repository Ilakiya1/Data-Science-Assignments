### 14. How is memory managed in Python? 


- Memory management in Python involves a private heap containing all Python objects and data structures. 
- The management of this private heap is ensured internally by the Python memory manager.
- The Python memory manager is responsible for allocating memory for Python objects and data structures and freeing up memory when it is not needed.

### 15) How to install Python on Windows and set path variables? 

- Download the latest version of python from its official website (https://www.python.org/downloads/).
- After downloading, install python by following the instructions. Check the option to add path environment variables.
- The steps to set the path variable as follows:
- Open start menu and search for 'Environment variables'.
- Click on 'Edit the system Environment variables'.
- In the advanced option, click on the 'Environment variables' button.
- Under system variable, find 'path' variable and click on edit.
- Click on 'New' and add path to the Python executable.
- Click ok and close all the windows and finally click 'apply' to make changes.
