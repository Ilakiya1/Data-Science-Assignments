### 1. Write a program to accept percentage from the user and display the grade according to the following criteria: 
Marks 
Grade 
>90 
A

```python
Marks = int(input("Enter your marks"))

if Marks > 90 and Marks <= 100:
    print("Grade A")
elif Marks > 80 and Marks <= 90:
    print("Grade B")
elif Marks >=60  and Marks <=80:
    print("Grade C")
else:
    print("Grade D")
```

### 2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria: 
Tax 
15% 
10% 
5% 
Cost Price(in Rs) 
>100000

```python
cost_price = int(input("Enter the price of a bike:"))

if cost_price > 100000:
    tax = cost_price * 0.15
    print("Road tax for the price of a bike greater than 100000 is {}".format(tax))
elif cost_price > 50000 and cost_price <= 100000:
    tax = cost_price * 0.10
    print("Road tax for the price of a bike between 50000 than 100000 is {}".format(tax))
elif cost_price < 50000:
    tax = cost_price * 0.05
    print("Road tax for the price of a bike lesser than 50000 is {}".format(tax))
else:
    print("No road tax found for the price of a bike")
```

### 3)  Accept any city from the user and display monuments of that city. 
City               Monument 
Delhi              Red Fort 
Agra               Taj Mahal 
Jaipur             Jal Mahal 

```python
city = input("Enter the city:")

if city == "Delhi" or city == "delhi":
    print("Monuments in {} are Red fort and Taj Mahal".format(city))
elif city == "Agra" or city == "agra":
    print("Monument in {} is Taj Mahal".format(city))
elif city == "Jaipur" or city == "jaipur":
    print("Monument in {} is Jal Mahal".format(city))
else:
    print("No Monuments found for {}".format(city))
```

### 4. Check how many times a given number can be divided by 3 before it is less than or equal to 10. 


```python
def div(num):
    count = 0
    while num > 10:
        num = num / 3
        count +=1
    return count

num = int(input("Enter a number:"))
div(num)
```

### 5. Why and When to Use while Loop in Python give a detailed description with example 


- While loop is used to execute a block of code until a certain condition is met.
- It will execute till the condition becomes False
- When the condition is always True, it will result in an infinite loop

Syntax for while loop is: 

```python
While condition:
   # block of code
```

```python
#For eg:
i = 1
while i <= 10:
    print(i)
    i = i+1
```

- while loops are useful when we are not sure about how many times to execute the loop
- There are several reasons why we need to use a while loop:
- If you want to repeat a specific task until a condition is met.
- If you want to read data from a collection or a file until there is no more data to read.
