### 1) What are the characteristics of the tuples? Is tuple immutable?

- Tuples are ordered collection of elements which can be accessed using indices. 
- Tuples can aslo contain different data types of elements
- Yes, Tuple is immutable, i.e the elements stored inside a tuple cannot be changed once they are created. The length of a tuple is also fixed and cannot be changed. Accessing elements can be faster than the list because of their immutable nature and fixed length.

- eg: Tuple with different data types



```python
Tuple1 = (1, 2.3, 'strings', (6,7), [1,2,4])
Tuple1
```

### 2) What are the two tuple methods in python? Give an example of each method. Give a reason why tuples have only two in-built methods as compared to Lists.

- The two tuple methods in python are count() and index()
- 1. Count(): This method returns the number of times a specified values occurs inside a tuple
- eg:

```python
num = (1,2,1,3,4,5,1)
num.count(1)
```

- Index() : This method returns index of first occurrence of a specified value in a tuple.
- eg:

```python
num1 = (1,2,3,4,5)
print(num.index(3))
```

- Tuple has only 2 methods when compared to list because the elements are immutable, i.e the elements cannot be changed and the main function of a tuple is to store values that should not be modified throughout the program

### 4) Explain the difference between the union() and update() methods for a set. Give an example of each method.

- Union(): This method in set returns a new set which contains all the elements from both the set, without modifying the original sets.
- For example

```python
set1 = {1,2,3}
set2 = {4,5,6}
set1.union(set2)
```

 - Update(): This method updates the original set by adding all the elements from the other set to it.
 - For example:

```python
set3 = {7,8,9}
set4 = {3,6,1}
set3.update(set4)
print(set3)
```

### Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display these three view objects for the given dictionary.
dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

- The three view objects in dictionaries in Python are
- 1) Keys()
- 2) Values()
- 3) Items()

- we can use these built-in methods to view the elements inside the dictionary

```python
dict1 =  {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

# Keys view object

keys = dict1.keys()
print(keys)

# Values view object

values = dict1.values()
print(values)

# Items view object

items = dict1.items()
print(items)
```
