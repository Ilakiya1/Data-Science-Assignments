### Q1. Which keyword is used to create a function? Create a function to return a list of odd numbers in the range of 1 to 25.

1. - '**def**' keyword is used to create a function in python.

```python
def odd_num():
    list1 = []
    for i in range(1,26):
        if i % 2 != 0:
            list1.append(i)
    return list1

odd_num()  
```

### Q2. Why '\*args' and '\*\*kwargs' is used in some functions? Create a function each for '\*args' and '\*\*kwargs' to demonstrate their use.

- **'\*args'** and **'\*\*kwargs'** are used to pass a variable number of arguments to a function
- **'\*args'** are non-keyword arguments
- **'\*\*kwargs'** are keyword arguments


```python
def add(*num):
    result = 0
    for i in num:
        result = result + i
    return result

add(5,5)
```

```python
def dict1(**data):
    for key,value in data.items():
        print('{} is {}'.format(key, value))

dict1(name = 'john', age = 30)
```

### Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator function.

- Generator function uses a generator object which can be iterated over a for loop or 'next()' keyword.
- In generator function, yield keyword is used instead of return keyword to return a value. 'yield' keyword is used because, it returns a value and then pauses its execution.


```python
def fib():
    a, b = 0,1
    while True:
        yield a
        a ,b = b, a + b

fibonacci = fib()
for i in range(10):
    print(next(fibonacci))
```

### Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the first 20 prime numbers.

```python
def prime():
    output = []
    for i in range(2,1000):
        prime = True
        for j in output:
            if j * j > i:
                break
            if i % j == 0:
                prime = False
                break
        if prime:
            output.append(i)
            yield i

prime_num = prime()
for i in range(20):
    print(next(prime_num))
```
