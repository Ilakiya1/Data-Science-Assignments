### Q3. Explain why the __init__() function is used. Give a suitable example.

- __init__ method is an in-built method that is used as a constructor in python
- This method is used to initialize the attributes of an object when it is first created.
- This method is automatically called when we create an instance of a class.

```python
class class1:
    def __init__(self, name, email_id, student_id):
        self.name = name
        self.email_id = email_id
        self.student_id = student_id
        
    def return_student_details(self):
        return self.name, self.email_id, self.student_id
```

```python
obj_class1 = class1("John", "john@gmail.com", 101)
obj_class1.name
```

```python
obj_class1.email_id
```

```python
obj_class1.return_student_details()
```

### Q4. Why self is used in OOPs?

- 'Self' is nothing but a pointer or reference to the object created by the class.
- The 'self' parameter allows the method to access and modify the attributes and behaviors of the object, which are defined in the class.

```python
# Example
class test:
    def welcome(self):
        print("Welcome to the class")
        
obj = test()
obj.welcome()
```

### Q5. What is inheritance? Give an example for each type of inheritance.

- A class which inherits properties from another class is known as Inheritance.
- The new class which is created is known as dervied class or subclass and the class existing class is known as base class or superclass
- There are several types of Inheritance in Python,
    1. Single Inheritance
    2. Multi-level Inheritance
    3. Multiple Inheritance

```python
# 1. Single Inheritance

class parent:
    def test_parent(self):
        print("This is the parent class")
        
class child(parent):
    pass
```

```python
# creating object for child class
obj_child = child()
obj_child.test_parent()
```

```python
# 2. Multi-level Inheritance

class class1:
    def test_class1(self):
        print("This is the class1")
        
class class2(class1):
    def test_class2(self):
        print("This is the class2")

class class3(class2):
    def test_class3(self):
        print("This is the class3")
```

```python
obj_class3 = class3()
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

```python
obj_class3.test_class3()
```

```python
# 3.Multiple Inheritance
class class1():
    def test_class1(self):
        print("This is the class1")
        
class class2():
    def test_class2(self):
        print("This is the class2")
        
class class3(class1, class2):
    pass
```

```python
obj_class3 = class3()
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```
